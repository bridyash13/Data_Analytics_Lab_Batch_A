```python
train_df = pd.read_csv ('../input/titanic/train.csv')
train_df = train_df

train_df.head(10)
```

```python
train_df.columns  = train_df.columns.str.lower()
train_df.columns
```

```python
train_df  = train_df.drop(columns=["passengerid", "name", "sibsp", "parch", "ticket", "cabin"])
print(train_df.shape,'\n', train_df.columns)
```

```python
train_df["age"].fillna(value=train_df["age"].mean(), inplace=True)
train_df["fare"].fillna(value=train_df["fare"].mean(), inplace=True)
train_df["embarked"].fillna(train_df["embarked"].mode()[0], inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
train_df['age'] = scaler.fit_transform(train_df[["age"]])*10
train_df['fare'] = scaler.fit_transform(train_df[["fare"]])*10
train_df
```

```python
train_df
```

```python
from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()
train_df['sex']= label_encoder.fit_transform(train_df['sex'])
train_df['embarked'] = label_encoder.fit_transform(train_df['embarked'])
train_df
```

```python
X = train_df.drop(columns="survived", axis=1)
Y= train_df["survived"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=.20, random_state=40)
```

**LOGISTIC REGRESSION**

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train, Y_train)
lr_predicted = lr.predict(X_train)
print(lr_predicted)
```

```python
from sklearn.metrics import accuracy_score

X_train_accuracy = accuracy_score(Y_train, lr_predicted)
print("Accuracy Score Of Linear Regression: ",X_train_accuracy)
```

**Random Forest Classifier**

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(X_train, Y_train)
rf_predicted = rf.predict(X_train)
print(rf_predicted)
```

```python
from sklearn.metrics import accuracy_score

X_train_accuracy = accuracy_score(Y_train, rf_predicted)
print("Accuracy Score Of Random Forest: ",X_train_accuracy)
```

**NAIVE BAYES**

```python
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb.fit(X_train, Y_train)
nb_predicted = nb.predict(X_train)
print(nb_predicted)
```

```python
from sklearn.metrics import accuracy_score

X_train_accuracy = accuracy_score(Y_train, nb_predicted)
print("Accuracy Score Of Naive Bayes: ",X_train_accuracy)
```

**LINEAR VECTOR QUANTIZATION**

```python
from sklearn import svm

svmm = svm.SVC()
svmm.fit(X_train, Y_train)
svm_predicted = svmm.predict(X_train)
print(svm_predicted)
```

```python
from sklearn.metrics import accuracy_score

X_train_accuracy = accuracy_score(Y_train, svm_predicted)
print("Accuracy Score Of SVM: ",X_train_accuracy)
```
