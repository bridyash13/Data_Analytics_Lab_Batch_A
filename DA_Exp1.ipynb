```python
import pandas as pd
import os
```

```python
raw_data = pd.read_excel('Customer.xlsx',nrows=10000)
raw_data
```

```python
raw_data.head()
```

```python
data = raw_data.drop(['EVR','CPI','CCI'], axis = 1)
data
```

```python
data.drop(data[(data['Job']=='Unknown') | (data['Loan_Outstanding']=='Unknown') | (data['Loan_Default_History']=='Unknown') | (data['Own_Housing']=='Unknown')].index,inplace=True)
data
```

```python
data[['Age','Call_Duration']].describe()
```

```python
print("\nEducation Levels:",data.Education.unique())
print("\nJobs:",data.Job.unique())
Ages = data.Age.unique()
Ages.sort()
print("\nAges:",Ages)
print("\nPeople with oustanding loans:",len(data[data['Loan_Outstanding']=='Yes']))
print("\nPeople who own a house:",len(data[data['Own_Housing']=='Yes']))
```

<h2>Data Visualization</h2>

```python
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,4))
sb.lineplot(x="Job", y="Convert", data=data)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(15,4))
sb.boxplot(x="Age",y="Convert",data=data)
plt.show()
```

```python
sb.histplot(data,x="Age",y="Call_Duration",kde=True)
```

```python
plt.figure(figsize=(15,4))
sb.boxplot(x="Call_Duration",y="Convert",data=data)
plt.show()
```
